```python
url= "https://www.ecok.edu/policies-and-handbooks"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
csv_file = open("ecu_policy_info.csv","w")

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['year', 'info'])

for lists in soup.find_all('div', class_="region region-content"):
  #table =lists.find('a',class_="btn btn-primary")
  #lists = soup.find('div', class_="region region-content")
  
  try:
    table =lists.find('a',class_="btn btn-primary")["href"]
    #name = table.split("/")
    name = table.split("/")[6]
    year = name.split("-")[0]
    ecu_link = f"https://www.ecok.edu/sites/default/files/website_files/Policies_and_Handbooks/Annual_Safety_Report{year}.pdf"
    
  except Exception as e:
    ecu_link = None

  print(ecu_link)
  print()

  csv_writer.writerow([year, ecu_link])

csv_file.close()
```
